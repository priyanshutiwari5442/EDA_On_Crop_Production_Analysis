## 3. ***Data Wrangling***

### Data Wrangling Code

```python
#Let see the percentage of null value present in production column
print((df['Production'].isnull().sum()*100)/len(df))

#So their are 1.51% data is null in production column
#Here we apply median to fill it in place of null value
production_median = df['Production'].median()
df['Production'].fillna(production_median,inplace = True)
```

####Add **<u>Crop Categories</u>** column ####

```python
#Lets create Crop categories list to add it further into Crop categories column
cereals = ['Wheat', 'Rice', 'Maize', 'Jowar', 'Bajra', 'Ragi', 'Barley', 'Small millets', 'Other Cereals&Millets', 'Korra', 'Samai']
pulses = ['Moong(Green Gram)', 'Urad, Arhar/Tur', 'Gram', 'Masoor', 'other misc. pulses', 'Other Rabi pulses', 'Peas & beans (Pulses)', 'Blackgram',
          'Lentil', 'Cowpea(Lobia)', 'Masoor', 'Horse-gram', 'Khesari', 'Guar seed', 'Moth', 'Rajmash Kholar', 'Ricebean(nagadal)']
oilseed = ['Groundnut', 'Sunflower', 'Castor seed', 'Sesamum', 'Linseed', 'Safflower', 'Rapeseed &Mustard', 'Niger seed', 'Oilseedstotal','other oilseeds']
Vegetables = ['Potato', 'Onion', 'Beans & Mutter(Vegetable)', 'Bhindi', 'Brinjal', 'Cucumber', 'Tomato', 'Cabbage', 'Peas (vegetable)',
              'Bottle Gourd', 'Garlic', 'Ginger', 'Carrot', 'Redish', 'Turnip', 'Cauliflower', 'Other Vegetables', 'Bitter Gourd', 'Drum Stick',
              'Jack Fruit', 'Snak Guard', 'Pump Kin', 'Ash Gourd', 'Beet Root', 'Lab-Lab', 'Ribed Guard', 'Yam', 'Colocosia']
Fruits = ['Citrus Fruit', 'Grapes', 'Mango', 'Orange', 'Papaya', 'Pome Fruit', 'Lemon', 'Pome Granet', 'Sapota', 'Pineapple', 'Apple', 'Peach',
          'Pear', 'Plums', 'Litchi', 'Ber', 'Other Fresh Fruits', 'Other Citrus Fruit', 'Water Melon', 'OtherDryFruit']
Cash_Crops = ['Cotton(lint)',' Tobacco', 'Jute', 'Rubber', 'Coffee', 'Tea', 'Mesta', 'Jute &mesta', 'Kapas']
Other_Crops = ['Arecanut', 'Cashewnut', 'Sugarcane', 'Sweet potato', 'Tapioca', 'Sannhamp', 'Varagu', 'Other fibres', 'Cond-spcs other',
               'Arcanut (Processed)', 'Atcanut (Raw)', 'Cashewnut Processed', 'Cashewnut Raw', 'Cardamom', 'Totalfoodgrain']
```

```python
#Create a function which help to create crop category
def crop_category(row):
  if row['Crop'] in cereals:
    return 'cereals'
  elif row['Crop'] in oilseed:
    return 'oilseed'
  elif row['Crop'] in Vegetables:
    return 'Vegetables'
  elif row['Crop'] in Fruits:
    return 'Fruits'
  elif row['Crop'] in Cash_Crops:
    return 'Cash_Crops'
  else:
    return 'Other_Crops'

#apply the above function to create Crop Categories column
df['Crop Categories'] = df.apply(crop_category,axis = 1)
```

####Add **<u>Yield</u>**  column####

```python
#Here we add a new column called yield
df['yield'] = df['Production']/df['Area']
```

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

#####Q1). What is the Total Production by year?#####

```python
year_production = df.groupby('Crop_Year')['Production'].sum().reset_index()
year_production

fig = px.bar(year_production, x = 'Crop_Year', y = 'Production', title = 'Production by Year',
                 labels = {'Crop_Year':'Crop Year', 'Production':'Total Production'})
fig.update_layout(showlegend = False)
fig.show()
```

#### Chart - 2

#####Q2). What are the most commonly grown crop in each state?#####

```python
#Here we group state name, crop and their count
state_crop = df.groupby(['State_Name','Crop'])['Crop'].count().reset_index(name = 'count')

#Here we write a function which help tp get most commonly grown crop
def max_count(Raw_state_crop_df):
  return Raw_state_crop_df.loc[Raw_state_crop_df['count'].idxmax()]

#And finally here our dataframe is ready
most_common_crop = state_crop.groupby('State_Name').apply(max_count).reset_index(drop = True)
most_common_crop
```

```python
#Here we group state name, crop and their count
state_crop = df.groupby(['State_Name','Crop'])['Crop'].count().reset_index(name = 'count')

#Here we write a function which help tp get most commonly grown crop
def max_count(Raw_state_crop_df):
  return Raw_state_crop_df.loc[Raw_state_crop_df['count'].idxmax()]

#And finally here our dataframe is ready
most_common_crop = state_crop.groupby('State_Name').apply(max_count).reset_index(drop = True)

fig = px.bar(most_common_crop, x='State_Name', y='count', color='Crop',
             title='Most Commonly Grown Crop in Each State',
             labels={'State_Name': 'State', 'count': 'Count of Crop', 'Crop': 'Crop Name'})

fig.update_layout(xaxis_tickangle=-45, showlegend=True)
fig.show()
```

#### Chart - 3

#####Q3). What is the highest crop production by season?

```python
season_production = df.groupby(['Crop', 'Season'])['Production'].sum().reset_index()
season_list =season_production.Season.unique()

# Filter the data for the winter season
for i in season_list:
  mask =season_production[season_production.Season == i]
  mask=mask.sort_values('Production', ascending=False)
  top_crops = mask.head(10)
  fig = px.bar(top_crops, x='Crop', y='Production', title= 'Highest Crop Production in {}'.format(i))
  fig.show()
```

#### Chart - 4

#####Q4).What is the distribution of crop production across different seasons?

```python
crop_seasons = df.groupby('Season')['Production'].count().reset_index()
colors = ['blue', 'yellowgreen', 'red', 'Purple', 'Pink', 'orange']

plt.figure(figsize = (10,8))
plt.pie(crop_seasons['Production'],autopct = '%1.2f%%',colors = colors)
plt.title('Percentage of Production for Crops in Each Season',bbox = {'facecolor':'grey','pad':0.2,'boxstyle':'round'})
plt.legend(crop_seasons['Season'])
plt.show()
```

#### Chart - 5

#####Q5).What is the change in production over the years in each state?

```python
state_production_per_year = df.groupby(['State_Name', 'Crop_Year'])['Production'].sum().reset_index()
state_production_per_year = state_production_per_year.sort_values('Crop_Year')
```

```python
fig = px.line(state_production_per_year, x='Crop_Year', y='Production', color='State_Name',
              title='Change in Production in Each State Over the Years')
fig.show()
```

#### Chart - 6

#####Q6).What is the avg yield per hectare for crop category?

```python
avg_yield_crop_category = df.groupby('Crop Categories')['yield'].mean().reset_index()
```

```python
colors = ['brown','blue', 'gray', 'lime', 'royalBlue', 'yellowgreen']

fig, ax = plt.subplots()

ax.pie(avg_yield_crop_category['yield'], labels=avg_yield_crop_category['Crop Categories'],
       autopct='%1.0f%%', colors = colors)

# Create a circle for the center
circle = plt.Circle((0, 0), 0.6, color='white')
ax.add_artist(circle)

# Equal aspect ratio ensures that the pie is drawn as a circle
ax.set_aspect('equal')
plt.title('Average Yield per Hectare for Crop Categories',bbox = {'facecolor':'darkKhaki','pad':0.2,'boxstyle':'round'})
plt.legend(avg_yield_crop_category['Crop Categories'])
plt.show()
```
